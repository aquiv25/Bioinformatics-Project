The columns are:

**idx**
The index of this profile, numbering each on in the file starting from 1.

**name**
The name of the profile.

**accession**
The optional accession of the profile, or "-" if there is none.

**nseq**
The number of sequences that the profile was estimated from.

**eff_nseq**
The effective number of sequences that the profile was estimated from, after HMMER applied an effective sequence number calculation such as the default entropy weighting.

**M**
The length of the model in consensus residues (match states).

**relent**
Mean relative entropy per match state, in bits. This is the expected (mean) score per consensus position. This is what the default entropy-weighting method for effective sequence number estimation focuses on, so for default HMMER3 models, you expect this value to reflect the default target for entropy-weighting.

**info**
Mean information content per match state, in bits. Probably not useful. Information content is a slightly different calculation than relative entropy.

**p relE**
Mean positional relative entropy, in bits. This is a fancier version of the per-match-state relative entropy, taking into account the transition (insertion/deletion) probabilities; it may be a more accurate estimation of the average score contributed per model consensus position.

**compKL**
Kullback-Leibler distance between the model's overall average residue composition and the default background frequency distribution. The higher this number, the more biased the residue composition of the profile is. Highly biased profiles can slow the HMMER3 acceleration pipeline, by causing too many nonhomologous sequences to pass the filters.

```python
import pandas as pd

def parse_tblout(filename):
    rows = []
    with open(filename) as f:
        for line in f:
            if line.startswith('#'): continue
            cols = line.split()
            if len(cols) < 10: continue
            rows.append({
                'protein_id': cols[0],
                'E-value':    float(cols[4]),
                'score':      float(cols[5]),
                'description': ' '.join(cols[18:])
            })
    return pd.DataFrame(rows)

threshold = 1e-5

for fname, label in [('ctcf_hits.txt', 'CTCF / zf-C2H2'),
                     ('smc_hits.txt',  'SMC'),
                     ('rad21_hits.txt','RAD21')]:
    df = parse_tblout(fname)
    sig = df[df['E-value'] < threshold].sort_values('E-value')
    print(f"\n=== {label} (E < {threshold}) : {len(sig)} hits ===")
    print(sig.to_string(index=False))
```

Мы выбрали следующие эпигенетические гены:

```python
from Bio import SeqIO

targets = {
    "DNMT1":  ["cytosine-5)-methyltransferase 1"],
    "TET3":   ["methylcytosine dioxygenase tet3"],
    "EZH2":   ["enhancer of zeste homolog 2"],
    "KMT2A":  ["lysine N-methyltransferase 2A", "myeloid/lymphoid"],
    "CHD1":   ["chromodomain-helicase-DNA-binding protein 1"],
    "BRD3":   ["bromodomain-containing protein 3"],
    "ADAR1":  ["RNA-specific adenosine deaminase isoform X1"],
    "UHRF1":  ["ubiquitin-protein ligase UHRF1", "E3 ubiquitin-protein ligase UHRF1"],
}

found = {}

for record in SeqIO.parse("GCF_001194135.2_ASM119413v2_protein.faa", "fasta"):
    full_text = (record.id + " " + record.description).lower()
    for gene_name, keywords in targets.items():
        if gene_name not in found:
            if any(kw.lower() in full_text for kw in keywords):
                found[gene_name] = record
                print(f"✓ {gene_name}: {record.id} — {record.description[:60]}")

print(f"\nНайдено: {len(found)}/{len(targets)}")


with open("block1_epigenetic_8genes_obimaculoides.fasta", "w") as out:
    SeqIO.write(list(found.values()), out, "fasta")
```

```python
import pandas as pd
from collections import defaultdict

rows = []
with open('/content/ZDNA_promoters-2.csv') as f:
    for line in f:
        cols = line.strip().split('\t')
        if len(cols) < 11: continue
        try:
            score = float(cols[3])
            gene = cols[7]
            desc = cols[10]
            rows.append((score, gene, desc))
        except:
            continue


gene_max = defaultdict(float)
gene_desc = {}
for score, gene, desc in rows:
    if score > gene_max[gene]:
        gene_max[gene] = score
        gene_desc[gene] = desc

keywords = ['transcription', 'zinc finger', 'histone', 'chromatin',
            'kinase', 'homeobox', 'factor', 'helicase', 'replication',
            'repair', 'recombination', 'cyclin', 'cohesin', 'SMC']

filtered = [(score, gene, gene_desc[gene])
            for gene, score in gene_max.items()
            if any(kw.lower() in gene_desc[gene].lower() for kw in keywords)
            and 'uncharacterized' not in gene_desc[gene].lower()
            and 'transfer RNA' not in gene_desc[gene]]

filtered.sort(reverse=True)
for s, g, d in filtered[:15]:
    print(f'{s:.3f}  {g}  {d}')
```

```python
genes = [
    "LOC106877206", "LOC106876744", "LOC106880212",
    "LOC106877680", "LOC106883024", "LOC106869367",
]

rows = []
with open('/content/ZDNA_promoters-2.csv') as f:
    for line in f:
        cols = line.strip().split('\t')
        if len(cols) < 11: continue
        if cols[7] in genes:
            rows.append(cols)

import pandas as pd
df = pd.DataFrame(rows, columns=['chr','zdna_start','zdna_end','score',
                                  'chr2','prom_start','prom_end','gene',
                                  'dot','strand','description'])
print(df[['gene','chr','prom_start','prom_end','strand','description']].drop_duplicates('gene').to_string(index=False))
```
